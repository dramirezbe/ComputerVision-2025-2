```python
import numpy as np
import matplotlib.pyplot as plt

#seed
np.random.seed(0)
```

### ðŸ“Œ Mini-proyecto: Visualizando experimentos

### Objetivo

Usar datos generados con **Numpy** (simulaciÃ³n de lanzamientos de dados) y graficarlos con **Matplotlib**, aplicando distintos tipos de grÃ¡ficos: histograma, barras y visualizaciÃ³n de matrices.


### Parte 1 â€“ SimulaciÃ³n con Numpy

Generamos 1000 lanzamientos de un dado (valores entre 1 y 6).

```python
throws = np.random.randint(1, 7, size=10000000)
```

### Parte 2 â€“ Histograma de resultados

Representamos la distribuciÃ³n de los lanzamientos.

```python
plt.hist(throws, bins=6, density=True, color='blue', edgecolor='black')
plt.title('Die Roll Histogram')
plt.xlabel('Die Face')
plt.ylabel('Count Throws')
plt.show()
```

### Parte 3 â€“ GrÃ¡fico de barras con probabilidades

Comparamos las probabilidades empÃ­ricas.

```python
values, counts = np.unique(throws, return_counts=True)
probabilities = counts / counts.sum()
plt.bar(values, probabilities, color='blue', edgecolor='black')
plt.xticks(values)
plt.xlabel('Die Face')
plt.ylabel('Probability')
plt.title('Die Roll Probabilities')
plt.show()
```

### Parte 4 â€“ Visualizar una matriz aleatoria con `imshow`

Creamos una matriz de valores aleatorios y la mostramos con un colormap.

```python
matrix = np.random.rand(100, 100)
plt.imshow(matrix, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Random 100x100 Matrix Visualization HOT Colormap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
